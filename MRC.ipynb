```python
from infer import tokenize_function, data_collator, extract_answer
from model.mrc_model import MRCQuestionAnswering
from transformers import AutoTokenizer
import nltk
nltk.download('punkt')
```

```python
model_checkpoint = "nguyenvulebinh/vi-mrc-large"
# model_checkpoint = "nguyenvulebinh/vi-mrc-base"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = MRCQuestionAnswering.from_pretrained(model_checkpoint)
```

```python
import json

# filepath = './assets/entity_types.json'
# # entity_type_list = None

# with open(filepath, 'r', encoding='utf-8') as file:
#     entity_type_list = json.load(file)

# print(entity_type_list)

file_faq = './assets/intent_1.json'
file_temp = './assets/template.json'
faq_conv = None
temp = None

with open(file_faq, 'r', encoding='utf-8') as file:
    faq_conv = json.load(file)
    print(faq_conv)

with open(file_temp, 'r', encoding='utf-8') as file:
    temp = json.load(file)
    print(temp)
```

```python
for conv in faq_conv:
    context = '. '.join(conv['message'])
    intent = conv['intent']
    if intent =='other':
        continue
    for question in temp[intent]:
        QA_input = {
            'question': question,
            'context': context
        }
        
        inputs = [tokenize_function(QA_input, tokenizer)]
        inputs_ids = data_collator(inputs, tokenizer)
        outputs = model(**inputs_ids)
        answer = extract_answer(inputs, outputs, tokenizer)

        print(answer)
```
